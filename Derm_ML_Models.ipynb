```python
import pandas as pd

# Load feature and target datasets
X = pd.read_csv("derm_data.csv")
y = pd.read_csv("derm_target.csv").squeeze() # convert dataframe to 1D array to better fit SVC models
```

.info() showed that all varibles are non-null other than 8 null age values, consistant with the information provided at the dataset sorce website.

```python
# Fill missing age values with the median to preserve distribution while avoiding NaNs
X['age'].fillna(X['age'].median(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Scale all features for better compatibility with SVM models
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Keep as a Dataframe
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Preview data after scaling
print(X_scaled.head()) # scaled
print(y.head()) # unchanged
```

```python
# Verify Mean of 0 and Std of 1
X_scaled.describe()
```

```python
# View outcome varible split to determine if stratifying is warrented in train test split
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

# Train-test split - #1

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
# Stratify train test split to ensure all classes are balanced in split
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Train Models - First Train_Test_Split

```python
# Import SVC Model - target data is categorical with labels 1-6
from sklearn.svm import SVC
```

```python
# Train first SVC model with a linear kernel and balanced class weight to compensate for class imbalance
svc_linear_1 = SVC(kernel='linear', C=1, class_weight='balanced')
svc_linear_1.fit(X_train, y_train)
```

```python
# Train second SVC model with a polynomial kernel and balanced class weight to compensate for class imbalance
svc_poly_1 = SVC(kernel='poly', C=1.5, gamma=0.5, class_weight='balanced')
svc_poly_1.fit(X_train, y_train)
```

```python
# Train third SVC model with a RBF kernel and balanced class weight to compensate for class imbalance
svc_rbf_1 = SVC(kernel='rbf', C=0.5, gamma='scale', class_weight='balanced')
svc_rbf_1.fit(X_train, y_train)
```

```python
# Import metrics libraries
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
)
```

```python
# Loop evalation metric print outs

# Define model name:object tuples
models = [
    ("Linear SVM 1", svc_linear_1),
    ("Polynomial SVM 1", svc_poly_1),
    ("RBF SVM 1", svc_rbf_1)
]
# Iterate loop for metrics per model
for name, model in models:
    print(f"\n=== {name} ===")
    y_pred = model.predict(X_test)

    # Accuracy
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
```

The trained support vector machine (SVM) classification models used linear, polynomial, and RBF kernels on a stratified train-test split of a dermatology dataset. Stratification was applied to ensure that all six target classes were proportionally represented in both the training and testing data.

- The Linear SVM achieved the highest accuracy at 97.30%, correctly classifying nearly all of the 74 test instances. It attained perfect precision and recall (1.00) for classes 1, 3, 5, and 6. Class 2 had a recall of 0.83, with two of its instances misclassified as class 4 — as confirmed by the confusion matrix.

- The Polynomial SVM had a lower accuracy of 90.54%. While precision remained high across most classes, recall for class 2 dropped to 0.67 and for class 6 to 0.75. The confusion matrix shows that the polynomial model had difficulty distinguishing between classes 2 and 4, misclassifying four instances of class 2 as class 4, along with one instance each of classes 3, 5, and 6. These patterns suggest that the polynomial kernel may have overfit the training data or misrepresented certain feature interactions, leading to increased misclassification.

- The RBF SVM achieved an accuracy of 95.95%, performing comparably to the linear model. It demonstrated strong classification across all classes, with minor dips in recall for class 2 (0.83) and class 3 (0.93). Most predictions were correct, and misclassifications were minimal, as shown in the confusion matrix.

This reinforced the value of stratified sampling in performance evaluation and led me to do one additional train-test split with a different random_state and reintroducing stratification.

Identify Top 10% Predictive Features

Using 'svc_linear_1' model, the top 10% the most predictive features using the .coef_ attribute will be selected. The linear SVMwas used because only linear models produce directly interpretable feature weights via .coef_. Polynomial and RBF kernels project the data into higher-dimensional spaces, where feature weights cannot be directly interpreted or translated into individual relationships.

```python
# Get the model coefficients
coefs = svc_linear_1.coef_
coefs
```

```python
# Convert to absolute values of coefs
import numpy as np
abs_coefs = np.abs(coefs)
```

```python
# Calculate the average imporatance across all classes
avg_importance = np.mean(abs_coefs, axis=0)
```

```python
# Extract the top 10% of features
num_top = int(0.10 * len(avg_importance))  # 10% of 34 = 3.4 → becomes 3
top_indices = np.argsort(avg_importance)[-num_top:]  # sorts and extracts the top 3
```

```python
# View top 3 as their original column names
top_features = X.columns[top_indices]
print("Top predictive features:", top_features.tolist())
```

```python
# Extract the top features into a new dataframe
X_top = X_scaled[top_features]
X_top.head()
```

```python
# Peform a new Train_Test_Split
X_train_top, X_test_top, y_train_top, y_test_top = train_test_split(
    X_top, y, test_size=0.2, random_state=33, stratify=y
)
```

```python
# Train first SVC model with a linear kernel and balanced class weight to compensate for class imbalance (Top 10% train_test_split)
svc_linear_top = SVC(kernel='linear', C=1, class_weight='balanced')
svc_linear_top.fit(X_train_top, y_train_top)
```

```python
# Train second SVC model with a polynomial kernel and balanced class weight to compensate for class imbalance (Top 10% train_test_split)
svc_poly_top = SVC(kernel='poly', C=1.5, gamma=0.5, class_weight='balanced')
svc_poly_top.fit(X_train_top, y_train_top)
```

```python
# Train third SVC model with a RBF kernel and balanced class weight to compensate for class imbalance (Top 10% train_test_split)
svc_rbf_top = SVC(kernel='rbf', C=0.5, gamma='scale', class_weight='balanced')
svc_rbf_top.fit(X_train_top, y_train_top)
```

```python
# Loop evalation metric print outs

# Define model name:object tuples
models_top = [
    ("Linear SVM Top 10%", svc_linear_top),
    ("Polynomial SVM Top 10%", svc_poly_top),
    ("RBF SVM Top 10%", svc_rbf_top)
]
# Iterate loop for metrics per model
for name, model in models_top:
    print(f"\n=== {name} ===")
    y_pred_top = model.predict(X_test_top)

    # Accuracy
    print(f"Accuracy: {accuracy_score(y_test_top, y_pred_top):.4f}")

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test_top, y_pred_top))

    # Confusion Matrix
    print("Confusion Matrix:")
    print(confusion_matrix(y_test_top, y_pred_top))
```

For the final set of models, only the top 10% most predictive features were used, identified from the linear SVM's learned coefficients, to assess how well the models could perform with a reduced feature set. This step aimed to test whether a simpler, more interpretable model could still achieve competitive accuracy. While feature selection can reduce overfitting and improve model explainability, the results here showed a significant drop in performance.

- The Linear SVM had an accuracy score of 66.22%. Performance declined sharply compared to the full feature model. Class 1 recall dropped to 0.48 despite perfect precision, meaning over half of class 1 samples were misclassified. Class 6 wasn't predicted at all, and class 3 showed both weak recall and precision.

- The Polynomial SVM reached an accuracy of 63.51%, slightly lower than the linear model. It struggled with class 1 and class 3 in particular, and class 6 had both low recall (0.25) and precision (0.33). Misclassifications were spread across multiple classes, suggesting overfitting and poor generalization.

- The RBF SVM had an accuracy score of 63.51% and mirrored the polynomial model's metrics and confusion matrix exactly. This is likely due to the limited features available to train on and the similarities between the polynomial and RBF models.

Overall, these results demonstrate that reducing the feature set to only the top 10% significantly weakened the models' ability to distinguish between all 6 classes. While feature selection is valuable in some contexts, this task requires more than 10% of features to provide reliable performance.

## Conclusion
This investigation applied supervised machine learning techniques to a dermatology dataset for multiclass classification. The dataset included 34 features and 6 distinct diagnostic classes. Initial preprocessing involved filling missing values, standardizing all features using StandardScaler, and ensuring the target variable was in a compatible format. Three Support Vector Machine (SVM) classifiers, Linear, Polynomial, and RBF, were trained and evaluated across three different train-test splits. The first split used stratification to maintain class balance; the second intentionally omitted stratification to examine the impact of uneven class distribution; and the third reintroduced stratification to validate model performance under more controlled conditions. Across these nine models, the Linear and RBF SVMs consistently outperformed the Polynomial SVM, with the highest accuracy in the unstratified split reaching 98.65% with the RBF model; and 97.30% in the stratified linear model.

Additonally, the top 10% most predictive features based on the linear model's coefficients were selected and used to retrain all three models. These reduced-feature models showed significantly lower performance (around 63-66% accuracy), especially struggling with underrepresented classes like class 6.

Overall, the results demonstrate that full-feature models offer improved generalization compared to reduced-feature models. Additionally, though the unstratified models achieved slightly higher accuracy scores, those results were misleading due to the absence of rare classes like class 6 in the test set. The Linear SVM, trained on the first stratified split, provided the most reliable and balanced performance across all six classes, making it the best overall model in terms of generalization and fairness.
